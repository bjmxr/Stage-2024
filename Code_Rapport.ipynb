```python
import numpy as np
import scipy as scp
import pylab as pyl
import pywt
import pandas as pd
import holoviews as hv
import param
import panel as pn
import matplotlib.pyplot as plt
from panel.pane import LaTeX
import requests
hv.extension('bokeh')
import warnings
warnings.filterwarnings('ignore')
from PIL import Image
from io import BytesIO


from holoviews import opts
import numpy as np
import scipy as scp
import pylab as pyl
import pywt
import pandas as pd
import holoviews as hv
import param
import panel as pn
import requests

from panel.pane import LaTeX
hv.extension('bokeh')
import warnings
warnings.filterwarnings('ignore')
from PIL import Image
from io import BytesIO
def GradientHor(x):
    y=x-np.roll(x,1,axis=1)
    y[:,0]=0
    return y
def GradientVer(x):
    y=x-np.roll(x,1,axis=0)
    y[0,:]=0
    return y
def DivHor(x):
    N=len(x[0])
    y=x-np.roll(x,-1,axis=1)
    y[:,0]=-x[:,1]
    y[:,N-1]=x[:,N-1]
    return y
def DivVer(x):
    N=len(x)
    y=x-np.roll(x,-1,axis=0)
    y[0,:]=-x[1,:]
    y[N-1,:]=x[N-1,:]
    return y
def Gradient(x):
    y=[]
    y.append(GradientHor(x))
    y.append(GradientVer(x))
    return y
def Div(y):
    x=DivHor(y[0])+DivVer(y[1])
    return x
```

```python
#DEBRUITAGE D'IMAGE



local=0
def chargeData(name):
    if local:
        res=np.array(Image.open("./Archive/img/Canaletto.jpeg")).astype(float)
    
    else:
        url='https://plmlab.math.cnrs.fr/dossal/optimisationpourlimage/raw/master/img/Canaletto.jpeg'
        response = requests.get(url)
        res=np.array(Image.open(BytesIO(response.content))).astype(float)
        
    return res
im=chargeData('Canaletto')
def ProjGradBouleInf(g,l):
    gh=g[0]
    gv=g[1]
    temp=g
    p0=gh-(gh-l)*(gh>l)-(gh+l)*(gh<-l)
    p1=gv-(gv-l)*(gv>l)-(gv+l)*(gv<-l)
    temp[0]=p0
    temp[1]=p1
    return temp
def fonction_score(p, y):
    divergence_p = Div(p)  # Calculer la divergence de p
    somme = divergence_p + y  # Ajouter y  la divergence
    norme_carree = np.linalg.norm(somme)**2  # Calculer la norme L2 au carr
    score = 0.5 * norme_carree  # Multiplier par 0.5
    return score
def GradientFonc(p, x):
    return np.array(Gradient(np.array(Div(p))+np.array(x)))

def FBDenoisingTV(y,l,step_size,Niter):
    s=[]
    x=0*y
    p=Gradient(x)
    for k in range(Niter):
        grad = np.array(GradientFonc(p, y))
        p=p - step_size * grad
        p=ProjGradBouleInf(p,l)
        #print(fonction_score(p,y))
        s.append(fonction_score(p,y))
    return y+Div(p),s

def add_gaussian_noise(image, mean=0, std=150):
    """
    Ajoute du bruit gaussien  une image en noir et blanc.
    
    :param image: Image en entre (numpy array).
    :param mean: Moyenne du bruit gaussien.
    :param std: cart-type du bruit gaussien.
    :return: Image bruite.
    """
    # Convertir l'image en float pour viter les problmes de dbordement/underflow
    image = image.astype(np.float32)
    
    # Gnrer le bruit gaussien
    noise = np.random.normal(mean, std, image.shape)
    
    # Ajouter le bruit  l'image
    noisy_image = image + noise
    
    # Clipper les valeurs pour qu'elles restent dans l'intervalle [0, 255]
    noisy_image = np.clip(noisy_image, 0, 255)
    
    # Convertir l'image bruite en uint8
    noisy_image = noisy_image.astype(np.uint8)
    
    return noisy_image



# Ajouter du bruit gaussien  l'image
noisy_image = add_gaussian_noise(im)
Niter=300
step_size=0.2
l=200
plt.imshow(im,cmap='gray')
plt.show()

plt.imshow(noisy_image,cmap='gray')
plt.show()

imf,s=FBDenoisingTV(noisy_image,l,step_size,Niter)

plt.imshow(imf,cmap='gray')
plt.show()
t=np.arange(Niter)
courbe1=hv.Curve((t,s),kdims='iterations',vdims='score FB')
hv.output(courbe1)
```

```python
#COMPRESSIVE SENSING

def extraire_colonnes(matrice, indices_colonnes):
    #Extrait les colonnes spcifies d'une matrice.
    matrice = np.array(matrice)  # Convertir en tableau numpy pour simplifier la manipulation
    return matrice[:, indices_colonnes]

def plot_f(size, f,nom):
    plt.figure(figsize=(30, 5))
    plt.bar(np.arange(size), f.squeeze(), color='blue')
    plt.xlabel('Indices')
    plt.ylabel(f'Valeur de {nom}')
    plt.title(f'Diagramme en btons de la fonction {nom}')
    plt.show()
    
def generate_f(size, p):
    # Gnration d'un vecteur avec `p` composantes non nulles
    if p > size:
        raise ValueError("Le nombre de composantes non nulles `p` ne peut pas dpasser la taille du vecteur `size`.")
    vector = np.zeros(size)  
    non_zero_indices = np.random.choice(size, p)  
    vector[non_zero_indices] = 1
    return vector.reshape(size,1), non_zero_indices  

def generate_vectors(size, M, mean=0, std=1):
    # Gnration de `M` vecteurs de taille `size` avec des valeurs alatoires suivant une distribution gaussienne
    vectors = np.random.normal(mean, std, (M, size))
    return np.array(vectors)  


def gradient_fonc(A, y, g):
    # Calcul du gradient de la fonction
    return np.dot(A.T,  np.array(np.dot(A, g))-np.array(y) ) # Retour du gradient


def fonction_score(y, A, g, lam):
    return 0.5*(np.linalg.norm(y-np.dot(A, g)))**2+lam*np.linalg.norm(g,ord=1)

def test(size, p, M, num_iterations):
    s=[]
    
    f, indices = generate_f(size, p)  # Gnration du vecteur `f` et des indices non nuls
    A = generate_vectors(size, M)  # Gnration de la matrice `A`
    
    l=[]
    for j in indices:
        Aj = A[:, j].reshape(-1, 1)  # Aj doit tre une colonne, donc on utilise reshape
        xj = f[j]
        B = Aj @ np.transpose(Aj)
        x = np.linalg.norm(np.linalg.inv(Aj @ Aj.T) * np.sign(xj))
        l.append(x)
        
   
    x2 = max(l) 
    lam =(1/x2)*0.1
    l2 = np.zeros((size,1))
    for k in indices:
        Ak = A[:, k].reshape(-1, 1)
        xk = f[k]
        z=xk-lam*np.linalg.norm(np.linalg.inv(Ak @ Ak.T) * np.sign(xk))
        l2[k]=z
    
    
    y = np.dot(A, f)  # Calcul du produit matrice-vecteur `y`
    g=np.zeros((size,1)) 
    plot_f(size,g,"g  l'itration 0")
    X=np.dot(A.T,A)
    eigenvalues, eigenvectors = np.linalg.eig(X)
    max_eigenvalue = np.max(eigenvalues)
    step=1/max_eigenvalue
    
   
    sl2=fonction_score(y,A,l2,lam)
    for i in range(num_iterations):  
        grad = gradient_fonc(A, y, g)
        g = g - step*grad  
        
        for j in range(g.shape[0]):  
               g[j, 0] = np.sign(g[j, 0]) * np.maximum(abs(g[j, 0]) - step*lam, 0) 
        
        s.append(fonction_score(y, A, g, lam))
        if i in range(1,7) :
            plot_f(size,g,f"g  l'itration {i}")
        if i in [10,15,19]:
            plot_f(size,g,f"g  l'itration {i}")
   
    
    # Affichage
    plt.show()
    plot_f(size,l2,'solution thorique')
    plot_f(size,f,'f')
    

  
    t = np.arange(num_iterations)
    title = f'Parcimonies: {p} Mesures: {M}'
    
    # Courbe des scores (non constante)
    plt.plot(t, s, color='tab:blue', label='Score FB')
    plt.xlabel('Iterations')
    plt.ylabel('Score FB')
    
    # Courbe constante avec les valeurs de l2
    l2_values = [sl2] * num_iterations
    plt.plot(t, l2_values, color='tab:red', linestyle='--', label='solution thorique')
    
    # Titre et lgende
    plt.title(title)
    plt.legend()
    
    
   
    return None 

size = 1500  # Taille des vecteurs
l = 5  # Nombre de configurations de `p`
c = 500  # Nombre de configurations de `M`
num_iterations=20

P=test(size,l,c,num_iterations)
 
```
